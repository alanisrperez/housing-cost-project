```python
import pandas as pd

# Load the .xls Excel file
KernCounty = pd.read_excel("Resources/BAKE$HWS.xls", engine='xlrd')

# Display the first few rows of the DataFrame
print (KernCounty)
```

```python
KernCounty.drop(['SORTORDER', 'BMYEAR', 'AREA', 'INCLUDES', 'BREAKS', 'PNCO', 'SS-NAICS'], axis=1, inplace=True)
```

```python
# Assuming your DataFrame is called df
columns_2010 = KernCounty.iloc[:, 1:13]  # Select the columns for 2010

# Calculate the row-wise average for these columns and round to 2 decimal places
KernCounty['2010_Average'] = columns_2010.mean(axis=1).round(2)

# Print the resulting DataFrame with the new '2010_Average' column
print(KernCounty[['TITLE', '2010_Average']])
```

```python
# Create an empty DataFrame to store the averages for each year
averaged_years_df = pd.DataFrame()

# Loop through each year from 2010 to 2024
for year in range(2010, 2025):
    # Adjust the column range based on your DataFrame structure
    # Replace '1:13' with the actual column range for each year
    start_col = (year - 2010) * 12 + 1  # Adjust column index logic as per your DataFrame structure
    end_col = start_col + 12
    
    # Select the columns for the current year
    columns_for_year = KernCounty.iloc[:, start_col:end_col]
    
    # Calculate the row-wise average for these columns and round to 2 decimal places
    KernCounty[f'{year}_Average'] = columns_for_year.mean(axis=1).round(2)
    
    # Add the year and its average column to the new DataFrame
    averaged_years_df[f'{year}_Average'] = KernCounty[f'{year}_Average']

# Add the 'TITLE' column to the new DataFrame
averaged_years_df['TITLE'] = KernCounty['TITLE']

# Reorder columns to put 'TITLE' at the front
averaged_years_df = averaged_years_df[['TITLE'] + [col for col in averaged_years_df.columns if col != 'TITLE']]

# Print or inspect the new DataFrame with averages from 2010 to 2024 and the TITLE column
print(averaged_years_df)
```
