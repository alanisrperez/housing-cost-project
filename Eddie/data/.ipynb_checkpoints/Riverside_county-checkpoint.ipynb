```python
import pandas as pd

# Load the .xls Excel file
RiversideCounty = pd.read_excel("RIVEHWS.xls", engine='xlrd')

RiversideCounty.head()
```

```python
RiversideCounty.drop(['SORTORDER', 'BMYEAR', 'AREA', 'INCLUDES', 'BREAKS', 'PNCO', 'SS-NAICS'], axis=1, inplace=True)

RiversideCounty.head()
```

```python
titles_df = RiversideCounty["TITLE"]
titles_df
```

```python
# Specify the number of spaces to filter titles with exactly 4 leading spaces
desired_indentation_level = 4

# Filter rows where the TITLE column starts with exactly 4 spaces
filtered_RiversideCounty = RiversideCounty[RiversideCounty['TITLE'].str.startswith(' ' * desired_indentation_level)]

# Create an empty DataFrame to store the averages for each year
averaged_years_df = pd.DataFrame()

# Loop through each year from 2010 to 2024
for year in range(2010, 2025):
    # Adjust the column range based on your DataFrame structure
    # Replace '1:13' with the actual column range for each year
    start_col = (year - 2010) * 12 + 1  # Adjust column index logic as per your DataFrame structure
    end_col = start_col + 12
    
    # Select the columns for the current year from the filtered DataFrame
    columns_for_year = filtered_RiversideCounty.iloc[:, start_col:end_col]
    
    # Calculate the row-wise average for these columns and round to 2 decimal places
    filtered_RiversideCounty[f'{year}_Average'] = columns_for_year.mean(axis=1).round(2)
    
    # Add the year and its average column to the new DataFrame
    averaged_years_df[f'{year}_Average'] = filtered_RiversideCounty[f'{year}_Average']

# Add the 'TITLE' column to the new DataFrame
averaged_years_df['TITLE'] = filtered_RiversideCounty['TITLE']

# Reorder columns to put 'TITLE' at the front
averaged_years_df = averaged_years_df[['TITLE'] + [col for col in averaged_years_df.columns if col != 'TITLE']]

# Print or inspect the new DataFrame with averages from 2010 to 2024 and the TITLE column
print(averaged_years_df)
```

```python
# Export the DataFrame to a CSV file
averaged_years_df.to_csv('Riverside_county.csv', index=False)  # Set index=False to avoid exporting the row indices
```
