```python
import pandas as pd
```

```python
avg_cost_by_county = pd.read_csv('../data/avg_cost_by_county.csv')
```

```python
avg_cost_by_county.head()
```

```python
avg_cost_by_county.columns
```

```python
home_df = avg_cost_by_county[avg_cost_by_county['State'] == 'CA']
home_df.head()
```

```python
avg_cost_2010up = pd.concat([home_df.iloc[:, :3], home_df.iloc[:, -175:]], axis=1)
avg_cost_2010up = avg_cost_2010up.drop('SizeRank', axis=1)
avg_cost_2010up.head(10)
```

```python
yr2010 = avg_cost_2010up.iloc[:,2:14]
yr2011 = avg_cost_2010up.iloc[:, 14:26]
yr2012 = avg_cost_2010up.iloc[:, 26:38]
```

```python
start_col = 2
num_cols = 12

dataframes = []
for i in range(start_col, len(avg_cost_2010up.columns), num_cols):
    end_col = min(i + num_cols, len(avg_cost_2010up.columns))
    new_df = avg_cost_2010up.iloc[:, i:end_col]
    dataframes.append(new_df)

# for idx, new_df in enumerate(dataframes):
#     print(f"DataFrame {idx + 1}")
#     print(new_df.head())

dataframes[2]
```

```python
#put all the years together
combined_df = pd.concat(dataframes, axis = 1)
combined_df.insert(0, 'RegionID', avg_cost_2010up.iloc[:,0])
combined_df.insert(1, 'County', avg_cost_2010up.iloc[:,1])
combined_df['County'] = combined_df['County'].str.replace(' County', '', regex=False)
combined_df
```

```python
combined_df.to_csv('housing_by_month.csv', index=False)
combined_df.to_json('housing_by_month.json', orient='records', lines=True)
```

```python
avg_dfs = []
start_year = 2010

for i in range(len(dataframes)):
    new_df = dataframes[i].iloc[:, 2:14].mean(axis=1)
    new_df = pd.DataFrame(new_df, columns= [start_year])
    start_year +=1
    avg_dfs.append(new_df)
```

```python
combined_yr_df = pd.concat(avg_dfs, axis=1)
combined_yr_df.insert(0, 'RegionID', avg_cost_2010up.iloc[:,0])
combined_yr_df.insert(1, 'County', avg_cost_2010up.iloc[:,1])
combined_yr_df['County'] = combined_yr_df['County'].str.replace(' County', '', regex=False)
combined_yr_df.head()
```

```python
combined_yr_df.to_csv('housing_by_year.csv', index=False)
combined_yr_df.to_json('housing_by_year.json', orient='records', lines=True)
```
