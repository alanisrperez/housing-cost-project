```python
import pandas as pd
```

```python
avg_cost_by_county = pd.read_csv('../data/avg_cost_by_county.csv')
```

```python
avg_cost_by_county.columns
```

```python
home_df = avg_cost_by_county[avg_cost_by_county['State'] == 'CA']
home_df.head()
```

```python
avg_cost_2010up = pd.concat([home_df.iloc[:, :3], home_df.iloc[:, -175:]], axis=1)
avg_cost_2010up = avg_cost_2010up.drop('SizeRank', axis=1)
avg_cost_2010up.head(10)
```

```python
yr2010 = avg_cost_2010up.iloc[:,2:14]
yr2011 = avg_cost_2010up.iloc[:, 14:26]
yr2012 = avg_cost_2010up.iloc[:, 26:38]
```

```python
start_col = 2
num_cols = 12
start_year = 2010

dataframes = []
for i in range(start_col, len(avg_cost_2010up.columns), num_cols):
    end_col = min(i + num_cols, len(avg_cost_2010up.columns))
    new_df = avg_cost_2010up.iloc[:, i:end_col]
    new_df.insert(0, 'RegionID', avg_cost_2010up.iloc[:,0])
    new_df.insert(1, 'RegionName', avg_cost_2010up.iloc[:,1])
    new_df.insert(2, 'Year', start_year)
    dataframes.append(new_df)
    start_year+=1

# for idx, new_df in enumerate(dataframes):
#     print(f"DataFrame {idx + 1}")
#     print(new_df.head())

dataframes[2]
```

```python
combined_df = pd.concat(dataframes, ignore_index=True)
combined_df
```

```python
combined_df.to_csv('housing_by_month.csv', index=False)
```
